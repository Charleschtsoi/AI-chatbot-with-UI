# Chapter 2

Football case: finding the most successful football manager in English professional football games

```python
import pandas as pd
# url for "football_managers_workfile.csv"
url = 'https://osf.io/t6dgh/download'
df = pd.read_csv(url)
df.head()
```

```python
# grouping based on team and manger using .groupby
group = df.groupby(["team", "manager_id", "manager_name"])
group.head()
```

```python
# create total number of games by managers using .agg
# size is a function counting the number of observations
# all value of div are E0
games = group.agg(manager_games=("div", "size"))
games.head()
```

```python
# grouping based on team and manger using .groupby
group = df.groupby(["team", "manager_id", "manager_name"])
group.head()
# create total number of points by managers using .agg
# adding all points of all games for the same manager in the same
# team: +3 for win; +1 for draw; 0 for loss
points = group.agg(manager_points=("points", "sum"))
points.head()
```

```python
# Join is combine the two datasets
# Default indices: team + manager_id + manager_name
win = games.join(points)
win.head()
```

```python
# calculate win ratio
win["win_ratio"] = (
win["manager_points"] / win ["manager_games"]
)
win.head()
```

```python
# Sort by win ratio from high to low
win = win.sort_values(["win_ratio"], ascending=False)
# top 10 managers
win.head(10)
```

```python
# remove caretakers
top = win[win["manager_games"]>18]
```

```python
# sort by win ratio again
top =top.sort_values("win_ratio", ascending=False)
top.head()
```

```python
# Select top 10 managers
top_10 = top.head(10).sort_values("win_ratio", ascending=True)
managers = top_10.index.get_level_values("manager_name")
points = top_10["win_ratio"]

import matplotlib.pyplot as plt
# Create bar chart
plt.figure(figsize=(10, 6))
plt.barh(managers, points)
plt.xlabel("Average Points per Game")
plt.ylabel("Manager")
plt.title("Top 10 Football Managers by Win Ratio")


# Show plot
plt.show()
```

```python
top_15 = win.head(15).sort_values("win_ratio", ascending=True)
managers = top_15.index.get_level_values("manager_name")
points = top_15["win_ratio"]

import numpy as np

# Identify caretakers (managers with 18 or fewer games)
# Create color array: red for caretakers, blue for non-caretakers
is_caretaker = top_15["manager_games"] <= 18
colors = np.where(is_caretaker, 'red', 'blue')

# Create bar chart
plt.figure(figsize=(10, 6))
plt.barh(managers, points, color=colors)
plt.xlabel("Average Points per Game)")
plt.ylabel("Manager")
plt.title("Top 15 Football Managers (Caretakers in Red)")
plt.tight_layout()

# Add legend
plt.legend(handles=[
    plt.Rectangle((0,0),1,1, color='blue', label='Non-Caretaker (>18 games)'),
    plt.Rectangle((0,0),1,1, color='red', label='Caretaker (â‰¤18 games)')
])

# Show plot
plt.show()
```
